```python
#importing the necessary  libraries
import numpy as np #for mathematical operation
import pandas as pd #for data manipulation
import sqlite3 #for server less database
```

```python
#Database Connection
# Establishing a connection to the SQLite database
db = "../01_data/movies_directors_dataset.sqlite"
con = sqlite3.connect(db)
cur = con.cursor()  # Creating a cursor object to execute SQL queries
#we are creating a database connection because sql lite database from now hold the movies and director table which will be analyse
```

# Insights code

```python
#insight 1 code:
# director who has directed most movies
cur.execute("""
           SELECT NAME,COUNT(ORIGINAL_TITLE)
           FROM MOVIES INNER JOIN DIRECTORS ON MOVIES.DIRECTOR_ID=DIRECTORS.ID 
           GROUP BY NAME 
           ORDER BY COUNT(ORIGINAL_TITLE) DESC LIMIT 1
           """)
director=cur.fetchall()
print(f"the director who directed the most movies is:{director[0][0]}")
```

```python
#insight 2 code:
#the director who is most bankable
cur.execute("""
           SELECT NAME,SUM(BUDGET) 
           FROM  MOVIES 
           INNER JOIN DIRECTORS ON MOVIES.DIRECTOR_ID=DIRECTORS.ID 
           GROUP BY NAME 
           ORDER BY SUM(BUDGET) DESC LIMIT 1""")
mbud=cur.fetchall()
print(f"the most bankable director is: {mbud[0][0]}")
```

```python
# insight 3 code:
# Top 5 Directors with the Highest Total Movie Budgets
cur.execute("""
    SELECT name, SUM(budget)
    FROM directors
    JOIN movies ON directors.id = movies.director_id
    GROUP BY name
    ORDER BY SUM(movies.budget) DESC
    LIMIT 5
""")
high_budget = cur.fetchall()
print("Top 5 Directors with the Highest Total Movie Budgets:", high_budget)
```

```python
# insight 4 code:
#Most Expensive Movie Directed by Each Director
cur.execute("""
    SELECT name, original_title, MAX(budget)
    FROM directors
    JOIN movies ON directors.id = movies.director_id
    GROUP BY name
    ORDER BY MAX(budget) DESC
""")
expense_movie = cur.fetchall()
print("Most Expensive Movie Directed by Each Director:", expense_movie)
```

```python
# insight 5 code:
# Most Popular Movies
cur.execute("""
    SELECT original_title, popularity
    FROM movies
    ORDER BY popularity DESC
    LIMIT 5;
""")
most_popular_movies = cur.fetchall()
print("Top 5 Most Popular Movies :",most_popular_movies)
```

```python
# insight 6 code:
# top 5 Movies with the Highest Vote Count
cur.execute("""
    SELECT original_title, vote_count
    FROM movies
    ORDER BY vote_count DESC
    LIMIT 5
""")
highest_vote_count_movies = cur.fetchall()
print(f"top 5 Movies with the Highest Vote Count are :{highest_vote_count_movies}")
```

```python
# insight 7 code:
#top 5 Directors with the Highest Average Vote Count
cur.execute("""
    SELECT name, AVG(vote_count)
    FROM 
    directors JOIN movies ON directors.id = movies.director_id
    GROUP BY name
    ORDER BY AVG(vote_count) DESC
    LIMIT 5
""")

directors_avg_vote_count = cur.fetchall()
print("top 5 Directors with the Highest Average Vote Count:",directors_avg_vote_count)
```
